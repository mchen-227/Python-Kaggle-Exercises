```python
X = df.drop('benign_0__mal_1',axis = 1).values
y = df['benign_0__mal_1'].values

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
X_train.shape #426 rolls and 30 features
```

# Create the model

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

#the last layer needs to be a sigmoid function because we need the model to do binary classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

# Train the model

```python
model.fit(x=X_train,y=y_train, epochs= 600, validation_data = (X_test,y_test))
```

### Plot out the "Loss"

```python
#we can plot out the training loss as well as the validation loss
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
#this is a typical case where the data has overfitted since so many epochs were run
```

## To Prevent Overfitting - we need early stop

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))

#the last layer needs to be a sigmoid function because we need the model to do binary classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor = 'val_loss',mode ='min', verbose = 1, patience = 25)
```

```python
model.fit(x=X_train,y=y_train, epochs= 600, validation_data = (X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15, activation='relu'))
model.add(Dropout(0.5))

#the last layer needs to be a sigmoid function because we need the model to do binary classification
model.add(Dense(1, activation='sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(x=X_train,y=y_train, epochs= 600, validation_data = (X_test,y_test),
         callbacks=[early_stop])
```

```python
model_loss_2 = pd.DataFrame(model.history.history)
```

```python
model_loss_2.plot()
```

```python
predictions = model.predict(X_test)
```

```python
y_pred = np.where(predictions > 0.5, 1,0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
