# Data Processing

```python
df = df.drop('date',axis=1)
df=df.drop('zipcode',axis=1)
```

```python
X = df.drop('price', axis=1).values
y = df['price'].values #it's important to specify that as TF doesn't work well wtih pandas data structure

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape #design neural network based on the shape of incoming data.
```

```python
#build the model
model = Sequential()

model.add(Dense (19, activation = 'relu'))
model.add(Dense (19, activation = 'relu'))
model.add(Dense (19, activation = 'relu'))
model.add(Dense (19, activation = 'relu'))

model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse') #define optimizer and define how loss function will be calculated
```

```python
#train the model
model.fit(x=X_train, y= y_train,
          validation_data = (X_test,y_test),batch_size = 128, epochs = 400)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot() #if you see MSE creeping up as Epochs increases, that is a sign of overfitting to the training data, 
#i.e. you see much larger loss on your validation data
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict (X_test)
```

```python
np.sqrt(mean_squared_error(y_test, predictions))
```

```python
plt.figure(figsize =(10,6))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
```
